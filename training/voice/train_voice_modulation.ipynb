## Voice modulation classifier (Librosa features)

This notebook shows how to train a simple classifier that predicts:
- `monotone`
- `expressive`

using features like pitch variance, energy variance, and spectral centroid.

### Data
Use your own labeled clips (10–20+). Save metadata in `training/student_samples/template.csv` or a separate CSV with paths + labels.


```python
import numpy as np
import pandas as pd
import librosa

# Example feature extractor for a single wav file

def extract_features(path: str, sr: int = 16000) -> dict:
    y, _ = librosa.load(path, sr=sr, mono=True)
    f0 = librosa.yin(y, fmin=80, fmax=320, sr=sr)
    f0 = f0[np.isfinite(f0)]
    pitch_var = float(np.std(f0) / max(1e-6, np.mean(f0))) if f0.size else 0.0
    rms = librosa.feature.rms(y=y)[0]
    energy_var = float(np.std(rms) / max(1e-6, np.mean(rms))) if rms.size else 0.0
    centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
    centroid_var = float(np.std(centroid) / max(1e-6, np.mean(centroid))) if centroid.size else 0.0
    return {"pitch_var": pitch_var, "energy_var": energy_var, "centroid_var": centroid_var}

extract_features("example.wav")
```
